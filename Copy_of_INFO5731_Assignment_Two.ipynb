```
#Remove noise, such as special characters and punctuations such as 

df['Extract_df']=df['Extract_df'].str.replace('[^\w\s]','')
df.head(100)
```

```
#Remove numbers
df['Extract_df']=df['Extract_df'].str.replace('[0-9]','')
df.head(100)
```

```
#Remove stopwords by using the stopwords list
from nltk.corpus import stopwords
import nltk

nltk.download('stopwords')                          
s_list = stopwords.words('english')
df['Extract_df'] = df['Extract_df'].apply(lambda x: ' '.join(item for item in x.split() if item not in s_list))
df[['Extract_df']].head(10)
```

```
# Lowercase all texts
df['Extract_df']=df['Extract_df'].str.lower()
df.head(10)
```

```
#Stemming
from nltk.stem import PorterStemmer

st = PorterStemmer()
df['Extract_df']=df['Extract_df'].apply(lambda x: ' '.join([st.stem(word) for word in x.split()]))
df.head(10)
```

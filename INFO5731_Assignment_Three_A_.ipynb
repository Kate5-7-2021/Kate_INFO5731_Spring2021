```
#(2) Calculate the probabilities for all the bigrams in the dataset by using the fomular 
#count(w2 w1) / count(w2). For example, count(really like) / count(really) = 1 / 3 = 0.33.

import pandas
collect= TextBlob (results)
bi_grams=TextBlob(results).ngrams(2)
#print(bi_grams)
case1={}                                        # create bigrams' dictionary
for gram in bi_grams:
  gram=tuple(gram)
  case1[gram]=case1.get(gram,0)+1
bi_gram_prob={}                                 # bigrams' probability calculations
for key, val in case1.items():
  bi_gram_prob[key]=val/collect.words.count(key[0])  
df = pandas.DataFrame.from_dict(bi_gram_prob, orient = 'index', columns=['Bi_gram_probability']) # create dataframe of the results
#df['Bi_gram_frequency']=[val for val in case1.values()]                # the result of running this line is another column with bigrams' frequencies
df.head(101)
```

```
import math
import numpy as np


def cosim (v1,v2):                  # calculate cosine similarity of two vectors
  x=v1.toarray()
  y=v2.toarray()
  return np.sum(x*y)/math.sqrt(np.sum(x*x)*np.sum(y*y))



i=1                         #calculate cosine similarity of the query
s_query_1 = {}
while i <=106:
  cosin_similar=cosim(tfs[106], tfs[i])
  cosin_similar = str(cosin_similar)
  s_query_1['my_doc'+str(i)] = cosin_similar
  i+=1
s_query_1_noNul= {k:float(v) for k,v in s_query_1.items() if v != 'nan'}              # remove Nulls
s_query_1 = sorted(s_query_1_noNul, key=s_query_1_noNul.get, reverse=True)
s_query_1_r = {key:rank for rank, key in enumerate(s_query_1,1)}

my_list_cosim = [s_query_1_noNul[my_doc] for my_doc in s_query_1]
r_df=pandas.DataFrame.from_dict (s_query_1_r,orient = 'index', columns = ['rank'])
r_df ['cosine similarity'] = my_list_cosim
r_df.head(100)

```
